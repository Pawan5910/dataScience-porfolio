```python
from transformers import pipeline
import torch
import torch.nn.functional as F
import pandas as pd
from datetime import date, timedelta, datetime
import tweepy
from textblob import TextBlob
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
#import the tweets scraped/collected
df_t = pd.read_csv('bit_tweet.csv')
```

```python
df_t
```

```python
#function for cleaning the tweets
def Cleaning(tw):
  tw = re.sub('#bitcoin', 'bitcoin', tw)
  tw = re.sub('#Bitcoin', 'Bitcoin', tw)
  tw = re.sub('#[A-Za-z0-9]+','',tw)
  tw = re.sub('\\n','',tw)
  tw = re.sub('https?:\/\/\S+', '', tw)
  return tw
```

```python
#converting the tweets to the string type 
df_t['Tweet'] = df_t['Tweet'].apply(str).astype("string")
```

```python
type(df_t['Tweet'])
```

```python
#Applying the cleaning function to the tweets
df_t['Cleaned_Tweets'] = df_t['Tweet'].apply(Cleaning)
```

```python
df_clean = df_t['Cleaned_Tweets']
```

```python
df_clean.head()
```

```python
clean_L = df_clean.tolist()
```

```python
clean_L
```

```python
#Calling the sentiment analysis model
classifier = pipeline("sentiment-analysis")
```

```python
#Applying the model to get the scores
results = classifier(clean_L)
```

```python
results
```

```python
df_scores = pd.DataFrame(results)
```

```python
#storing the results obtained
df_scores = pd.read_csv('HF_results.csv')
```

```python
df_scores['score'] = df_scores['score'].apply(float).astype("float")
```

```python
#Calculating the Positive Scores 
df_scores['Positive_Score'] = np.where(df_scores['label'] == 'POSITIVE', df_scores['score'], 1- df_scores['score'])                 
```

```python
#Calculating the Negative Scores
df_scores['Negative_Score'] = np.where(df_scores['label'] == 'NEGATIVE', df_scores['score'], 1- df_scores['score']) 
```

```python
#Merge the scores with the tweets
new_df = pd.concat([df_t,df_scores],axis=1)
```
